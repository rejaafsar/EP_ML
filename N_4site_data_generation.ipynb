```python
from scipy import linalg
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X_train=np.array([[0,0,0]])
Y_train=np.array([[0,0]])
#X_train5=np.array([[0,0,0,0,0]])
Y_train5=np.array([[0,0,0,0]])
i=0
```

```python
############## Non EP points generation ######################

p=0.1
q=1.01
d=.1
m=0
n=0
q=0.1


#print("p","\t","q","\t","d","\t","rig1","\t","rig2","\t","rig3","\t","rig4","\n")
for m in range(20000):
   # q=1.0+(p**2)/d
    #if q<5:
        #i=i+1
        H=np.array([[1j*d,p,0,0],[p,1j,q,0],[0,q,-1j,p],[0,0,p,-1j*d]])
        Evals,LEvecs,REvecs=linalg.eig(H,left=True, right=True)
        rig1=(np.dot(np.conjugate(LEvecs[:,0]),REvecs[:,0]))/(np.dot(np.conjugate(REvecs[:,0]),REvecs[:,0]))
        rig2=(np.dot(np.conjugate(LEvecs[:,1]),REvecs[:,1]))/(np.dot(np.conjugate(REvecs[:,1]),REvecs[:,1]))
        rig3=(np.dot(np.conjugate(LEvecs[:,2]),REvecs[:,2]))/(np.dot(np.conjugate(REvecs[:,2]),REvecs[:,2]))
        rig4=(np.dot(np.conjugate(LEvecs[:,3]),REvecs[:,3]))/(np.dot(np.conjugate(REvecs[:,3]),REvecs[:,3]))
        
        #print(p,"\t",q,"\t",d,"\t",np.abs(rig1),"\t",np.abs(rig2),"\t",np.abs(rig3),"\t",np.abs(rig4),"\n")
       # print("m=",m,"\n")
        arr_x=np.array([[p,q,d]])
        arr_y=np.array([[-np.log(np.abs(rig1)),-np.log(np.abs(rig4))]])
        arr_y5=np.array([[-np.log(np.abs(rig1)),-np.log(np.abs(rig2)),-np.log(np.abs(rig3)),-np.log(np.abs(rig4))]])
        X_train=np.append(X_train,arr_x,axis=0)
        Y_train=np.append(Y_train,arr_y,axis=0)
        Y_train5=np.append(Y_train5,arr_y5,axis=0)
        p=round(np.random.uniform(0,5), 3)
        d=round(np.random.uniform(0,5), 3)
        q=round(np.random.uniform(0,5), 3)
    
```

```python
Data=np.concatenate((X_train,Y_train5),axis=1)
df = pd.DataFrame(Data)
df
```
