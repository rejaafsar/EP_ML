```python
# keras:
from tensorflow.keras import Sequential # Sequential is the neural-network class
from tensorflow.keras.layers import Dense # Dense is the standard network layer
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
import datetime
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'\Users\iamaf\OneDrive\Desktop\OneDrive - Indian Institute of Science\reja_awadhesh\NN_data\data_2site')
df
```

```python
df[df.iloc[:,2]<0.8]
```

```python
df = df.sample(frac=1).reset_index(drop=True)
df
```

```python
train, test = train_test_split( df, test_size=0.1, random_state=4)

X_train = train[['0','1']].values
Y_train = train[['2']].values

X_test = test[['0','1']].values
Y_test = test[['2']].values
```

```python
# X_test
```

```python
# Building the Neural_Network

Net=Sequential() # creating a neural network!
Net.add(Dense(16,input_shape=(2,),activation="relu")) # first hidden layer: 20 neurons (and 1 input neuron!)
#Net.add(Dense(4,activation="relu")) # second hidden layer: 10 neurons
Net.add(Dense(4,activation="relu")) # second hidden layer: 10 neurons
Net.add(Dense(1,activation="linear")) # output layer: 1 neuron
# Compile network: (randomly initialize weights, choose advanced optimizer, set up everything!)
Net.compile(loss='mean_squared_error',
              optimizer='adam') # adam is adaptive and works better than normal gradient descent

# Does the training
Net.fit(X_train, Y_train, epochs=200, batch_size=32, validation_data=(X_test, Y_test),shuffle=True )
```

```python
Net.save(r"\Users\iamaf\OneDrive\Desktop\OneDrive - Indian Institute of Science\reja_awadhesh\EP_ML\Model_2site.h5")
```
