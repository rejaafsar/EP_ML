```python
from scipy import linalg
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
N=7
X_data=np.array([[N,0,0]])
Y_data=np.array([np.zeros(N)])

H=np.zeros((N,N))
rig=np.zeros(N)


for m in range(20000):
    t1=0.000001 + round(np.random.uniform(0,1), 5) /100000
    t2=  round(np.random.uniform(0,1), 5)# /100000  
    for i in range(N):
        for j in range(N):
            if i==j+1:
                H[i,j]=t1
            if i==j-1:
                H[i,j]=t2
    Evals,LEvecs,REvecs=linalg.eig(H,left=True, right=True)            
    for i in range(N):
        rig[i]=(np.dot(np.conjugate(LEvecs[:,i]),REvecs[:,i]))/(np.dot(np.conjugate(REvecs[:,i]),REvecs[:,i]))
    arr_x=np.array([[N,t1,t2]])
   # arr_y= np.array([rig])
    arr_y= np.array([-np.log(rig)])
    X_data=np.append(X_data,arr_x,axis=0)
    Y_data=np.append(Y_data,arr_y,axis=0)
    

Data=np.concatenate((X_data,Y_data),axis=1)
df = pd.DataFrame(Data)
df
```

```python
train, test = train_test_split( fdf, test_size=0.1, random_state=4)

X_train = train[['0','1','2']].values
Y_train = train[['X']].values

X_test = test[['0','1','2']].values
Y_test = test[['X']].values
```

```python
# keras:
from tensorflow.keras import Sequential # Sequential is the neural-network class
from tensorflow.keras.layers import Dense # Dense is the standard network layer
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import tensorflow as tf
import datetime
```

```python
################# Building the Neural_Network


Net=Sequential() # creating a neural network!
Net.add(Dense(12,input_shape=(3,),activation="relu")) # first hidden layer: 20 neurons (and 1 input neuron!)
#Net.add(Dense(50,activation="relu")) # second hidden layer: 10 neurons
#Net.add(Dense(100,activation="relu")) # second hidden layer: 10 neurons
Net.add(Dense(4,activation="relu"))
Net.add(Dense(1,activation="linear")) # output layer: 1 neuron


# Compile network: (randomly initialize weights, choose advanced optimizer, set up everything!)
# Net.compile(loss='mean_squared_error',
#               optimizer=Adam(learning_rate=0.001)

 # adam is adaptive and works better than normal gradient descent
Net.compile(loss='mean_squared_error',
              optimizer='adam')
```

```python
# Does the training
Net.fit(X_train, Y_train, epochs=200, batch_size=32, validation_data=(X_test, Y_test))
```
